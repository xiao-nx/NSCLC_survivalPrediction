```python
import numpy as np 
import pandas as pd
```

```python
# 指定文件路径
file_path = "lung_cancer_1.xlsx"

# 读取数据集 
data = pd.read_excel(file_path, sheet_name='Sheet1')
data.head()
```

```python
case_ID = data['Patient'].values
case_index = []
for ID in case_ID:
    ID = ID[4:-4]
    case_index.append(int(ID))

print(case_index)
```

```python
# 获取指定行
matched_cases = data.iloc[case_index]
matched_cases
```
